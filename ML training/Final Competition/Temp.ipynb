```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def f(x):
    return x**3

sample_size = 10000
x_values = np.random.uniform(np.power(3, 1/3), np.power(2, 1/3), sample_size)

y_values = f(x_values)

sns.kdeplot(y_values, fill=True, color='blue', linewidth=2)

plt.title('Плотность распределения f(x)')
plt.xlabel('Значение f(x)')
plt.ylabel('Плотность')

plt.show()
```
